```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import numpy as np
```

# **Поиск простых чисел: решето Эратосфена, решето Аткина, решето Сундарама**

На рассмотренных объемах решето Эратосфена является наиболее быстрым методом поиска простых чисел до определенного предела, поскольку время, затраченное на поиск, растет медленнее по сравнению с другими методами.  

```python
x = [100, 1000, 10000, 50000, 100000, 500000, 1000000]
# n = [25, 168, 1229, 5133, 9592, 41538, 78498] - количество простых чисел в соответствующем интервале x

y_e = [0, 0, 0.003, 0.025, 0.034, 0.196, 0.400, 4,368]
y_a2 = [0, 0.001, 0.006, 0.043, 0.053, 0.275, 0.527]
y_s = [0, 0, 0, 0.026, 0.056, 0.332, 0.675]

plt.plot(x, y_e)
plt.plot(x, y_a2)
plt.plot(x, y_s)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')

```

# **Поиск простого числа по индексу**


```python
x1 = [5, 10, 15, 20, 50, 100, 1000, 5000, 10000, 50000, 100000]
y1 = [8.63, 19.3, 33.1, 50.9, 167, 434, 9150, 79900, 215000, 2480000, 7200000]

x2 = [20, 50, 100, 1000, 5000, 10000, 50000, 100000]
y2 = [19.4, 137, 299, 6300, 26200, 123000, 581000, 1280000]

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')

x = np.linspace(0,100000,5000)
y = (x * np.sqrt(x))/4.5
yy = (x * np.log(np.log(x)))*5

plt.plot(x, y)
plt.plot(x, yy)
```
